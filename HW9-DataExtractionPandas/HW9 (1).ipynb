## Question 8
Create a dataframe, _total_population_, showing the total population (across all groups) of each country for all the years.  Display the dataframe.

```python
# Question 8
total_population = population.groupby("Country Name").sum()
total_population
```

## Question 9
Create a dataframe, _top_five_, showing the total population of the 5 most populous coutries in 2021.   
Even though the countries are ranked based on their 2021 population, the dataframe must include data for all years.

Display the dataframe.


```python
# Question 9
top_five = total_population.sort_values(by= "2021", ascending=False).head(5)
top_five
```

## Question 10

# Population by Age

## Question 14
Create a dataframe, active_population, that shows for each country, the total 'working age' population (Population ages 15-64).
Use a set for the working age groups.
Use the _query_ method to select the relevant rows.
Display the dataframe after you create it.

```python
# Question 14
working_age = {"Population ages 15-64, male", "Population ages 15-64, female"}
active_population = population.query("Group in @working_age").groupby("Country Name").sum()
active_population
```

## Question 16
Create a pivot table, _pivot_, showing for each country the population in each group for 2021.  The column headers should be the groups. Display the pivot table.

In this part we'll use the land.csv dataset which is based on the Land area (sq. km) downloaded from the World Bank data catalog.

```python
# load the land.csv file
land = pd.read_csv("land.csv")
land.head(10)
```

## Question 18
Combine the land dataframe with the _total_population_ dataframe that you created in Question 8.  Name the resulting dataframe _countries_.  Display the _countries_ dataframe.

```python
# Question 18
countries = pd.merge(left=land, right=total_population, left_on="Country Name", right_on="Country Name")
countries
```

## Question 19
We define the population density as the total population of a country in 2021 divided by the land area in sq. km.
Add a column, _Density_,  to the countries dataframe that shows the population density of each country.  Display the _countries_ dataframe.

```python
# Question 19
countries["Density"] = countries["2021"] / countries["Area"]
countries
```

## Question 20
What are the 10 countries with the highest population density (with the highest one first)?
Your result should show a dataframe with the country names and the corresponding densities.  There is no need to show the yearly population totals, the Area or the country codes.

```python
# Question 20
countries.loc[:,["Country Name","Density"]].sort_values(by= "Density", ascending=False).head(10)
```
